```python
import pandas as pd
df = pd.read_csv('Housing.csv')
df
```

```python
df['mainroad'] = df['mainroad'].astype('category')
df['mainroad'] = df['mainroad'].cat.codes
df
```

```python
df['guestroom'] = df['guestroom'].astype('category')
df['guestroom'] = df['guestroom'].cat.codes

#basement
df['basement'] = df['basement'].astype('category')
df['basement'] = df['basement'].cat.codes

#hotwaterheating
df['hotwaterheating'] = df['hotwaterheating'].astype('category')
df['hotwaterheating'] = df['hotwaterheating'].cat.codes

#airconditioning
df['airconditioning'] = df['airconditioning'].astype('category')
df['airconditioning'] = df['airconditioning'].cat.codes

#prefarea
df['prefarea'] = df['prefarea'].astype('category')
df['prefarea'] = df['prefarea'].cat.codes 

#furnishingstatus 
df['furnishingstatus'] = df['furnishingstatus'].astype('category')
df['furnishingstatus'] = df['furnishingstatus'].cat.codes

df
```

#  Predicting Price from all

price is Dependent(Y) Feature, remaining are Independent (x1,x2,x3..)
y = m1x1 + m2x2 + m3x3 + ..... + c

```python
x = df.drop(columns = 'price')
x
# training features
```

```python
y = df['price']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
c = lr.intercept_
c
```

```python
m = lr.coef_
m
```

```python
y_pred_train = lr.predict(x_train)
y_pred_train
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_train,y_pred_train)
plt.xlabel("acctual")
plt.ylabel("predicted")
plt.show()
```

```python
from sklearn.metrics import r2_score

r2_score(y_train,y_pred_train)
```

```python
y_pred_test = lr.predict(x_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred_test)
plt.xlabel("acctual")
plt.ylabel("predicted")
plt.show()
```

```python
r2_score(y_test,y_pred_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.scatter(y_test, y_pred_test, label='Data points')
coefficients = np.polyfit(y_test, y_pred_test, 1)
regression_line = np.polyval(coefficients, y_test)
plt.plot(y_test, regression_line, color='red', label='Regression line')

plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.legend()
plt.show()
```
